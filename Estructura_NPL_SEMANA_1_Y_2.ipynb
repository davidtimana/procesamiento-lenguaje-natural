**Lematizaci√≥n:** devuelve las ra√≠ces (ej: usan ‚Üí usar, estudiantes ‚Üí estudiante).

**An√°lisis b√°sico de texto en espa√±ol con spaCy y pandas**

Este ejemplo muestra c√≥mo procesar un texto en espa√±ol utilizando spaCy para obtener informaci√≥n ling√º√≠stica de cada palabra (token). Se extraen datos como la forma original, su versi√≥n en min√∫sculas, el lema (forma base), si es una stopword y si corresponde a puntuaci√≥n. Luego, con pandas, se organiza la informaci√≥n en una tabla para facilitar el an√°lisis y la visualizaci√≥n.


En este ejemplo se procesa un texto en espa√±ol con spaCy y se construye una tabla usando pandas.
De cada palabra (token) se extrae:

**Token** ‚Üí el texto original.

**Normalizado** ‚Üí la palabra en min√∫sculas.

**Lema** ‚Üí forma base de la palabra (ejemplo: m√©dicos ‚Üí m√©dico).

**Es stopword** ‚Üí indica con True o False si la palabra es una stopword (muy
frecuente y poco informativa, como el, de, y).

**Es puntuaci√≥n** ‚Üí indica con True o False si el token corresponde a un signo de puntuaci√≥n.

***En la tabla resultante, los valores True significan que el token cumple con esa condici√≥n, mientras que False indica que no.***

```python
# ===========================================
# 1. Importar librer√≠as y cargar modelo
# ===========================================
import spacy
import pandas as pd

# Cargar modelo en espa√±ol
nlp = spacy.load("es_core_news_sm")

# ===========================================
# 2. Texto de ejemplo (puedes cambiarlo)
# ===========================================
texto = "Los m√©dicos recomiendan realizar actividad f√≠sica tres veces por semana para mejorar la salud cardiovascular, seg√∫n la ONU."

# Procesar el texto con spaCy
doc = nlp(texto)

# ===========================================
# 3. Construcci√≥n de tabla
# ===========================================
datos = []
for token in doc:                           # Itera sobre cada token (palabra, puntuaci√≥n, etc.) del documento procesado por spaCy
    datos.append({                          # Agrega un diccionario con informaci√≥n del token a la lista 'datos'
        "Token": token.text,                # Texto original del token tal como aparece
        "Normalizado": token.text.lower(),  # Versi√≥n en min√∫sculas (√∫til para an√°lisis de frecuencia)
        "Lema": token.lemma_,               # Forma base de la palabra (ej: "corriendo" ‚Üí "correr")
        "Es stopword": token.is_stop,       # Indica si el token es una palabra vac√≠a (ej: "y", "de", "el")
        "Es puntuaci√≥n": token.is_punct     # Indica si el token es un signo de puntuaci√≥n
    })


# Crear DataFrame con pandas
df = pd.DataFrame(datos)

# Mostrar tabla sin filas duplicadas de puntuaci√≥n
df = df[df["Es puntuaci√≥n"] == False].reset_index(drop=True)
df
```

**Procesamiento de frases en espa√±ol con spaCy: tokenizaci√≥n, normalizaci√≥n, limpieza y lematizaci√≥n**

```python
# Importa la librer√≠a spaCy, usada para procesamiento de lenguaje natural (tokenizaci√≥n, lematizaci√≥n, etc.)
import spacy
# Importa la librer√≠a pandas, utilizada para manejar y analizar datos en estructuras tipo DataFrame (tablas)
import pandas as pd


# Cargar modelo en espa√±ol
nlp = spacy.load("es_core_news_sm")

# Dataset de frases de prueba
frases = [
    "Los estudiantes de ingenier√≠a usan Python para proyectos de inteligencia artificial en la universidad.",
    "El mercado financiero mostr√≥ una ca√≠da significativa debido a la inflaci√≥n y la inestabilidad pol√≠tica en la regi√≥n.",
    "La inteligencia artificial est√° revolucionando la forma en que las empresas gestionan datos y procesos de negocio."
]

# Funci√≥n para procesar cada frase
def procesar_texto(texto):
    doc = nlp(texto)
    tokens = [token.text for token in doc]
    normalizados = [token.text.lower() for token in doc]
    limpios = [token.text.lower() for token in doc if not token.is_stop and not token.is_punct]
    lemas = [token.lemma_ for token in doc if not token.is_stop and not token.is_punct]
    return {
        "Texto original": texto,
        "Tokenizaci√≥n": tokens,
        "Normalizaci√≥n": normalizados,
        "Limpieza": limpios,
        "Lematizaci√≥n": lemas
    }

# Procesar todas las frases
resultados = [procesar_texto(frase) for frase in frases]

# Convertir a DataFrame para verlo como tabla
df = pd.DataFrame(resultados)

# Mostrar tabla
import IPython.display as display
display.display(df)
```

# **An√°lisis l√©xico y morfol√≥gico robusto en espa√±ol con spaCy**

```python
# ===========================================
# 0) Carga robusta de spaCy + modelo espa√±ol
# ===========================================
def _ensure_spacy_es():
    try:
        import spacy
        try:
            nlp = spacy.load("es_core_news_sm")
            return nlp
        except Exception:
            # Modelo no disponible: lo descargamos
            import sys, subprocess
            subprocess.check_call([sys.executable, "-m", "spacy", "download", "es_core_news_sm"])
            nlp = spacy.load("es_core_news_sm")
            return nlp
    except ImportError:
        # spaCy no est√°: lo instalamos y luego cargamos el modelo
        import sys, subprocess
        subprocess.check_call([sys.executable, "-m", "pip", "install", "-q", "spacy==3.7.5"])
        subprocess.check_call([sys.executable, "-m", "spacy", "download", "es_core_news_sm"])
        import spacy
        nlp = spacy.load("es_core_news_sm")
        return nlp

nlp = _ensure_spacy_es()

# ===========================================
# 1) Oraciones de prueba (ajusta o agrega las tuyas)
#    Incluye variaciones para ver POS y morfolog√≠a
# ===========================================
sentencias = [
    "Los m√©dicos recomiendan realizar actividad f√≠sica tres veces por semana.",
    "El ni√±o corri√≥ r√°pido, pero las ni√±as est√°n corriendo ahora.",
    "Cantar me relaja; cantaba todos los d√≠as y ayer cant√© en p√∫blico.",
    "La m√©dica y los m√©dicos atendieron a los pacientes.",
    "Me siento feliz, aunque antes era infeliz; valoro la felicidad."
]

# ===========================================
# 2) Funci√≥n de an√°lisis ‚Üí DataFrame por oraci√≥n
# ===========================================
import pandas as pd

def analizar_texto(texto):
    doc = nlp(texto)
    filas = []
    for token in doc:
        if token.is_space:
            continue
        filas.append({
            "Token": token.text,
            "Lema": token.lemma_,
            "POS": token.pos_,                 # categor√≠a gramatical (l√©xico)
            "Morfolog√≠a": token.morph.to_dict(),  # rasgos morfol√≥gicos (g√©nero, n√∫mero, tiempo, etc.)
            "Stopword": token.is_stop
        })
    df = pd.DataFrame(filas, columns=["Token", "Lema", "POS", "Morfolog√≠a", "Stopword"])
    return df

# ===========================================
# 3) Mostrar una tabla por oraci√≥n
# ===========================================
from IPython.display import display, HTML

for i, s in enumerate(sentencias, start=1):
    print(f"\nüìù Oraci√≥n {i}: {s}")
    df = analizar_texto(s)
    display(df)
```
