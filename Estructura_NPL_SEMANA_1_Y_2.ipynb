**Análisis básico de texto en español con spaCy y pandas**

Este ejemplo muestra cómo procesar un texto en español utilizando spaCy para obtener información lingüística de cada palabra (token). Se extraen datos como la forma original, su versión en minúsculas, el lema (forma base), si es una stopword y si corresponde a puntuación. Luego, con pandas, se organiza la información en una tabla para facilitar el análisis y la visualización.


En este ejemplo se procesa un texto en español con spaCy y se construye una tabla usando pandas.
De cada palabra (token) se extrae:

**Token** → el texto original.

**Normalizado** → la palabra en minúsculas.

**Lema** → forma base de la palabra (ejemplo: médicos → médico).

**Es stopword** → indica con True o False si la palabra es una stopword (muy
frecuente y poco informativa, como el, de, y).

**Es puntuación** → indica con True o False si el token corresponde a un signo de puntuación.

***En la tabla resultante, los valores True significan que el token cumple con esa condición, mientras que False indica que no.***

```python
# ===========================================
# 1. Importar librerías y cargar modelo
# ===========================================
import spacy
import pandas as pd

# Cargar modelo en español
nlp = spacy.load("es_core_news_sm")

# ===========================================
# 2. Texto de ejemplo (puedes cambiarlo)
# ===========================================
texto = "Los médicos recomiendan realizar actividad física tres veces por semana para mejorar la salud cardiovascular, según la ONU."

# Procesar el texto con spaCy
doc = nlp(texto)

# ===========================================
# 3. Construcción de tabla
# ===========================================
datos = []
for token in doc:                           # Itera sobre cada token (palabra, puntuación, etc.) del documento procesado por spaCy
    datos.append({                          # Agrega un diccionario con información del token a la lista 'datos'
        "Token": token.text,                # Texto original del token tal como aparece
        "Normalizado": token.text.lower(),  # Versión en minúsculas (útil para análisis de frecuencia)
        "Lema": token.lemma_,               # Forma base de la palabra (ej: "corriendo" → "correr")
        "Es stopword": token.is_stop,       # Indica si el token es una palabra vacía (ej: "y", "de", "el")
        "Es puntuación": token.is_punct     # Indica si el token es un signo de puntuación
    })


# Crear DataFrame con pandas
df = pd.DataFrame(datos)

# Mostrar tabla sin filas duplicadas de puntuación
df = df[df["Es puntuación"] == False].reset_index(drop=True)
df
```

**Procesamiento de frases en español con spaCy: tokenización, normalización, limpieza y lematización**

```python
# Importa la librería spaCy, usada para procesamiento de lenguaje natural (tokenización, lematización, etc.)
import spacy
# Importa la librería pandas, utilizada para manejar y analizar datos en estructuras tipo DataFrame (tablas)
import pandas as pd


# Cargar modelo en español
nlp = spacy.load("es_core_news_sm")

# Dataset de frases de prueba
frases = [
    "Los estudiantes de ingeniería usan Python para proyectos de inteligencia artificial en la universidad.",
    "El mercado financiero mostró una caída significativa debido a la inflación y la inestabilidad política en la región.",
    "La inteligencia artificial está revolucionando la forma en que las empresas gestionan datos y procesos de negocio."
]

# Función para procesar cada frase
def procesar_texto(texto):
    doc = nlp(texto)
    tokens = [token.text for token in doc]
    normalizados = [token.text.lower() for token in doc]
    limpios = [token.text.lower() for token in doc if not token.is_stop and not token.is_punct]
    lemas = [token.lemma_ for token in doc if not token.is_stop and not token.is_punct]
    return {
        "Texto original": texto,
        "Tokenización": tokens,
        "Normalización": normalizados,
        "Limpieza": limpios,
        "Lematización": lemas
    }

# Procesar todas las frases
resultados = [procesar_texto(frase) for frase in frases]

# Convertir a DataFrame para verlo como tabla
df = pd.DataFrame(resultados)

# Mostrar tabla
import IPython.display as display
display.display(df)
```

# **Análisis léxico y morfológico de un texto en español con spaCy**

```python

# ===========================================
# 2. Importar librerías y cargar modelo
# ===========================================
import spacy

# Cargar modelo en español
nlp = spacy.load("es_core_news_sm")

# ===========================================
# 3. Texto de ejemplo
# ===========================================
texto = "Los médicos recomiendan realizar actividad física tres veces por semana."

# Procesar texto
doc = nlp(texto)

# ===========================================
# 4. Mostrar análisis léxico y morfológico
"""
:<15 → el texto ocupará 15 espacios, alineado a la izquierda.
:<20 → el texto ocupará 20 espacios, alineado a la izquierda.
:<30 → el texto ocupará 30 espacios, alineado a la izquierda.
"""
# ===========================================
print(f"{'Token':<15} {'Léxico (POS)':<20} {'Morfología':<30}")
print("-"*70)
for token in doc:
    if not token.is_punct:  # ignorar signos de puntuación
        print(f"{token.text:<15} {token.pos_:<20} {token.morph}")
```

# **Análisis léxico y morfológico robusto en español con spaCy**

```python
# ===========================================
# 0) Carga robusta de spaCy + modelo español
# ===========================================
def _ensure_spacy_es():
    try:
        import spacy
        try:
            nlp = spacy.load("es_core_news_sm")
            return nlp
        except Exception:
            # Modelo no disponible: lo descargamos
            import sys, subprocess
            subprocess.check_call([sys.executable, "-m", "spacy", "download", "es_core_news_sm"])
            nlp = spacy.load("es_core_news_sm")
            return nlp
    except ImportError:
        # spaCy no está: lo instalamos y luego cargamos el modelo
        import sys, subprocess
        subprocess.check_call([sys.executable, "-m", "pip", "install", "-q", "spacy==3.7.5"])
        subprocess.check_call([sys.executable, "-m", "spacy", "download", "es_core_news_sm"])
        import spacy
        nlp = spacy.load("es_core_news_sm")
        return nlp

nlp = _ensure_spacy_es()

# ===========================================
# 1) Oraciones de prueba (ajusta o agrega las tuyas)
#    Incluye variaciones para ver POS y morfología
# ===========================================
sentencias = [
    "Los médicos recomiendan realizar actividad física tres veces por semana.",
    "El niño corrió rápido, pero las niñas están corriendo ahora.",
    "Cantar me relaja; cantaba todos los días y ayer canté en público.",
    "La médica y los médicos atendieron a los pacientes.",
    "Me siento feliz, aunque antes era infeliz; valoro la felicidad."
]

# ===========================================
# 2) Función de análisis → DataFrame por oración
# ===========================================
import pandas as pd

def analizar_texto(texto):
    doc = nlp(texto)
    filas = []
    for token in doc:
        if token.is_space:
            continue
        filas.append({
            "Token": token.text,
            "Lema": token.lemma_,
            "POS": token.pos_,                 # categoría gramatical (léxico)
            "Morfología": token.morph.to_dict(),  # rasgos morfológicos (género, número, tiempo, etc.)
            "Stopword": token.is_stop
        })
    df = pd.DataFrame(filas, columns=["Token", "Lema", "POS", "Morfología", "Stopword"])
    return df

# ===========================================
# 3) Mostrar una tabla por oración
# ===========================================
from IPython.display import display, HTML

for i, s in enumerate(sentencias, start=1):
    print(f"\n📝 Oración {i}: {s}")
    df = analizar_texto(s)
    display(df)
```
